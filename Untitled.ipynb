```python
csv1.dropna(axis=1) #  Along axis
```

```python
csv1.dropna(how="all")  # All :- remove row with all NaN , ANY:- remove row with any NaN 
```

```python
csv1.dropna(subset=["col4"])  # remove NaN along column
```

```python
csv1.dropna(inplace=True)  # remove NaN values row and col
```

```python
csv1.dropna(thresh=1)  # Remove the values with have only 1 NaN as 1 is given as parameter
```

# Merge and Concat DataFrames

# Merge

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,5],"C":[21,22,23,24,25]})

#  TO Merge the dataframes one part should be common (like ID)
pd.merge(var1,var2,on="A")  # On is used as common key

```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[21,22,23,24,25]})

pd.merge(var1,var2,how="right",indicator =True)  # left,outer, right
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6],"B":[21,22,23,24,25]})

pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","python"))

#suffixes gives name to column and right_index and left_index merge the table as Ax Ay and Bx By
```

# Concate

```python
sr1 = pd.Series([1,2,3])
sr2 = pd.Series([11,12,13,14])

pd.concat([sr1,sr2])
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2,3,4,6],"C":[21,22,23,24,25]})

pd.concat([var1,var2],axis=1)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2],"C":[21,22]})

pd.concat([var1,var2],axis=1)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2],"C":[21,22]})

pd.concat([var1,var2],axis=1,join="inner")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2],"C":[21,22]})

pd.concat([var1,var2],axis=1,join="outer")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 = pd.DataFrame({"A":[1,2],"C":[21,22]})

pd.concat([var1,var2],axis=0,keys=["d1","d2"])
```

# Join and Append DataFrame

# Join

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=["a","b","c","d","e"])
var2 = pd.DataFrame({"C":[1,2,3,4,6],"D":[21,22,23,24,25]},index=["a","b","d","c","e"])

var1.join(var2)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=["a","b","c","d","e"])
var2 = pd.DataFrame({"C":[1,2,3],"D":[21,22,23]},index=["a","b","d"])

var2.join(var1)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=["a","b","c","d","e"])
var2 = pd.DataFrame({"C":[1,2,3],"D":[21,22,23]},index=["a","b","d"])

var2.join(var1,how="right")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=["a","b","c","d","e"])
var2 = pd.DataFrame({"C":[1,2,3],"D":[21,22,23]},index=["a","b","d"])

var2.join(var1,how="left")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=["a","b","c","d","e"])
var2 = pd.DataFrame({"C":[1,2,3],"D":[21,22,23]},index=["a","b","d"])

var2.join(var1,how="outer")  # Outer == UNION and Inner == INTERSECTION
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=["a","b","c","d","e"])
var2 = pd.DataFrame({"C":[1,2,3],"D":[21,22,23]},index=["a","b","d"])

var2.join(var1,how="inner")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]},index=["a","b","c","d","e"])
var2 = pd.DataFrame({"C":[1,2,3],"B":[21,22,23]},index=["a","b","d"])

var2.join(var1,how="outer",lsuffix="_12",rsuffix="_12") # when column name are save it gives error changing column name in run time to _12
```

# Pivot Table and Melt Function

# MELT

```python
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15],"C":[21,22,23,25,26]})

var1
```

```python
pd.melt(var1)
```
